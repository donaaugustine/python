```
#pgm to demonstrate basic data type in python
x = 20     #int
print(x)
print(type(x))
print("\n")

x = 20.5   #float
print(x)
print(type(x))
print("\n")

x = 1j     #complex
print(x)
print(type(x))
print("\n")

x = "hello"  #string 
print(x)
print(type(x))
print("\n")

x = ["apple", "orange", "mango"]       #list
print(x)
print(type(x))
print("\n")

x = ("Python", "Java", "C")       #tuple
print(x)
print(type(x))
print("\n")

x = {"Name" : "Dona", "Age" : 20}  #dictionary
print(x)
print(type(x))
print("\n")

x = True              #bool
print(x)
print(type(x))
print("\n")

x = b"Hello"         #bytes
print(x)
print(type(x))
print("\n")

x = {"papaya", "mango", "orange"}   #set
print(x)
print(type(x))
print("\n")

x = frozenset({"papaya", "mango", "orange"})    #frozenset
print(x)
print(type(x))
```

```
#Fibonacci sequence
n = int(input("How many fibonacci numbers you want ? "))

n1, n2 = 0, 1
i = 0

if n <= 0:
   print("Please enter a positive integer")
elif n == 1:
   print("Fibonacci sequence : " , n1)
else:
   print("Fibonacci sequence:")
   while i < n:
       print(n1)
       nth = n1 + n2
     
       n1 = n2
       n2 = nth
       i += 1  
```

```
#Reverse of a number
n = int(input("Enter a number : "))
rev = 0
while n > 0:
  r = n % 10
  rev = rev*10+r
  n = n // 10
print("Reverse of number : " , rev)  
```

```
#check whether the palindrome or not
n = int(input("Enter a number : "))
num = n
rev = 0
while n > 0:
  r = n % 10
  rev = rev*10+r
  n = n // 10
if num == rev:
  print("Palindrome")
else:
  print("Not palindrome")    
```
